```python
import os
import shutil

# Define source directories
source_folders = {
    "positive": "/content/drive/MyDrive/TharuWorks/Roboflow_Input/Negative_Extract/train",
    "negative": "/content/drive/MyDrive/TharuWorks/Roboflow_Input/Positive_Extract/train"
}

# Define destination folder
destination_folder = "/content/drive/MyDrive/TharuWorks/Dataset/train"

# Create organized folders
os.makedirs(os.path.join(destination_folder, "images"), exist_ok=True)
os.makedirs(os.path.join(destination_folder, "depths"), exist_ok=True)
os.makedirs(os.path.join(destination_folder, "labels"), exist_ok=True)
os.makedirs(os.path.join(destination_folder, "metadata"), exist_ok=True)

# Loop through both Positive and Negative datasets
for label, source_folder in source_folders.items():
    for subfolder in os.listdir(source_folder):
        subfolder_path = os.path.join(source_folder, subfolder)

        if os.path.isdir(subfolder_path):  # Check if it's a directory
            for file in os.listdir(subfolder_path):
                file_path = os.path.join(subfolder_path, file)
                file_name, file_ext = os.path.splitext(file)

                # Construct new name with parent folder (positive/negative) and subfolder prefix
                new_name = f"{label}_{subfolder}_{file}"

                # Rename and move images
                if file.endswith((".jpg", ".png")):
                    if "depth" in file:
                        shutil.move(file_path, os.path.join(destination_folder, "depths", new_name))
                    else:
                        shutil.move(file_path, os.path.join(destination_folder, "images", new_name))

                # Rename and move labels
                elif file.endswith(".txt"):
                    shutil.move(file_path, os.path.join(destination_folder, "labels", new_name))

                # Rename and move metadata
                elif file.endswith(".json"):
                    shutil.move(file_path, os.path.join(destination_folder, "metadata", new_name))

print("✅ Dataset successfully organized into 'Organized_Dataset'.")
```
