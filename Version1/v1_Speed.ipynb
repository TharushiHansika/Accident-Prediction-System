```python
import pandas as pd  # Import pandas
import numpy as np

# Load bounding box data
df = pd.read_csv("/content/drive/MyDrive/TharuWorks/Frames/Negative_Extract/bounding_box_data.csv")

# Assuming each frame corresponds to a consistent time interval
time_interval = 1 / 30  # Replace with actual frame rate, e.g., 1/30 for 30 FPS

vehicle_speeds = []

for i in range(1, len(df)):
    # Get current and previous bounding box data
    prev_row = df.iloc[i - 1]
    curr_row = df.iloc[i]

    if prev_row['class_id'] == curr_row['class_id']:  # Check same vehicle class
        # Calculate the center points of the bounding boxes
        prev_center = ((prev_row['x_min'] + prev_row['x_max']) / 2, (prev_row['y_min'] + prev_row['y_max']) / 2)
        curr_center = ((curr_row['x_min'] + curr_row['x_max']) / 2, (curr_row['y_min'] + curr_row['y_max']) / 2)

        # Euclidean distance between centers
        distance_px = np.sqrt((curr_center[0] - prev_center[0])**2 + (curr_center[1] - prev_center[1])**2)

        # Estimate speed in pixels per second
        speed = distance_px / time_interval
        vehicle_speeds.append({
            "frame": curr_row['frame'],
            "speed_px_per_s": speed,
            "vehicle_class": curr_row['class_id']
        })

# Convert results to DataFrame for easier analysis
speed_df = pd.DataFrame(vehicle_speeds)
speed_df.to_csv("/content/drive/MyDrive/TharuWorks/Frames/Negative_Extract/vehicle_speeds.csv", index=False)

print("Speed estimation completed. Results saved to CSV.")
```
