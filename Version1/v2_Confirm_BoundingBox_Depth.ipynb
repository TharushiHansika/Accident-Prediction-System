```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Define dataset paths
dataset_root = "/content/drive/MyDrive/TharuWorks/Dataset/train"  # Adjust path
images_dir = os.path.join(dataset_root, "images")  # RGB images
depths_dir = os.path.join(dataset_root, "depths")  # Depth maps
depth_labels_dir = os.path.join(dataset_root, "depth_labels")  # Annotated labels

# New folder to save annotated images
annotated_images_dir = os.path.join(dataset_root, "annotated_images")
os.makedirs(annotated_images_dir, exist_ok=True)

# Choose a test file to visualize (Change this to a filename in your dataset)
test_filename = "negative_w10_106_frame_0_jpg.rf.b06634c70a47a905bb15f20dbc729086.txt"  # Change this to a valid label file

# Load corresponding image and depth map
image_filename = test_filename.replace(".txt", ".jpg")  # Assuming RGB images are .jpg
depth_filename = test_filename.replace(".txt", ".jpg")  # Assuming depth maps are also .jpg

image_path = os.path.join(images_dir, image_filename)
depth_path = os.path.join(depths_dir, depth_filename)
label_path = os.path.join(depth_labels_dir, test_filename)

# Load the image
image = cv2.imread(image_path)
if image is None:
    print(f"❌ Error loading RGB image: {image_filename}")
else:
    # Load depth image (grayscale)
    depth_map = cv2.imread(depth_path, cv2.IMREAD_GRAYSCALE)

    if depth_map is None:
        print(f"❌ Error loading depth image: {depth_filename}")
    else:
        # Load bounding box labels
        if not os.path.exists(label_path):
            print(f"❌ Label file not found: {label_path}")
        else:
            with open(label_path, "r") as f:
                lines = f.readlines()

            # Get image dimensions
            img_height, img_width, _ = image.shape

            # Draw bounding boxes
            for line in lines:
                parts = line.strip().split()
                if len(parts) < 6:
                    continue  # Skip invalid lines

                class_id, x_center, y_center, width, height, avg_depth = map(float, parts)

                # Convert YOLO format (normalized) to pixel values
                x_min = int((x_center - width / 2) * img_width)
                y_min = int((y_center - height / 2) * img_height)
                x_max = int((x_center + width / 2) * img_width)
                y_max = int((y_center + height / 2) * img_height)

                # Ensure bounding box is within valid bounds
                x_min, y_min = max(0, x_min), max(0, y_min)
                x_max, y_max = min(img_width - 1, x_max), min(img_height - 1, y_max)

                # Draw rectangle on the image
                color = (0, 255, 0)  # Green color for bounding box
                cv2.rectangle(image, (x_min, y_min), (x_max, y_max), color, 2)

                # Overlay depth value on the image
                text = f"Depth: {avg_depth:.2f}"
                cv2.putText(image, text, (x_min, y_min - 5),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

            # Save the final annotated image
            annotated_image_path = os.path.join(annotated_images_dir, image_filename)
            cv2.imwrite(annotated_image_path, image)
            print(f"✅ Annotated image saved: {annotated_image_path}")

            # Show the image with bounding boxes
            plt.figure(figsize=(10, 6))
            plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for Matplotlib
            plt.title(f"Bounding Boxes & Depth on {image_filename}")
            plt.axis("off")
            plt.show()
```
